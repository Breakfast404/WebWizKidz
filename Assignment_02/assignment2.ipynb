```python
def nextPage():
    try:
        driver.find_element_by_xpath('//button[@class="paginationWidget-next"]').click()
        time.sleep(1)
    except:
        "last page"
    
def getColumns():
    table = driver.find_element_by_xpath('//table[@class="stats_table data_grid"]')
    col_headers = table.find_elements(By.TAG_NAME, "th")
    labels = [x.text for x in col_headers]
    return labels
def getTableFromPage():
    table = driver.find_element_by_xpath('//table[@class="stats_table data_grid"]')
    trs = table.find_elements(By.TAG_NAME, "tr")
    totalList = []
    print('we did the first thing')
    n = 1
    for x in trs:
        if n % 10 == 0:
            print(str(n) + ' elements have been download')
        row = x.find_elements(By.TAG_NAME, "td")
        totalList += [[x.text for x in row]]
        n += 1
    return totalList[1:]
    
def getLabels():
    table = driver.find_element_by_xpath('//table[@class="stats_table data_grid"]')
    text = table.text
    new = text.split('\n')[0].split(' ')
    return [x.encode('utf-8').decode('utf-8') for x in new]

def getTable():
    global teamAbrev
    table = driver.find_element_by_xpath('//table[@class="stats_table data_grid"]')
    text = table.text
    new = text.split('\n')
    new = new[1:]
    count = 0
    for x in new:
        for team in teamAbrev:
            if x.find(team) != -1:
                ind = x.find(team)
        half = x[:ind].split(' ')
        otherhalf = x[ind:].split(' ')
        string = ''
        for item in half[1:]:
            string += item
        new[count] = [half[0]] + [string] + otherhalf
        count += 1
    return new

```
